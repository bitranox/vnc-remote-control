# vnc-remote-control - Quickstart

`vnc-remote-control` is a layout-aware VNC/RFB remote-control CLI: type, key,
click, screenshot, OCR, against any VNC/RFB server. It is a pure client; you
install it on the machine you drive from, and it connects out to a target's
VNC/RFB port. This notebook installs the published package and shows the CLI
surface. The commands that actually drive a guest (`type`, `click`,
`screenshot`, ...) need a reachable VNC server and a `--port`, so here we only
show `--version`, `info`, and `--help`.

## The command surface

Every VNC subcommand takes the global `--host` (default `127.0.0.1`) and
`--port`. Authentication is password-only (no username): `--password` or the
`VNC_REMOTE_CONTROL_PASSWORD` environment variable. `--delay-scale` slows every
input event for a sluggish guest.

## Driving a guest (needs a running VNC server)

With a reachable RFB server on, say, port 5901 you would run, for example:

```bash
vnc-remote-control --port 5901 screenshot /tmp/s.png   # prints: resolution: WxH
vnc-remote-control --port 5901 click-text "Next"        # OCR + click by label
vnc-remote-control --port 5901 type "hello" --enter
```

OCR commands (`ocr`, `click-text`) additionally require the `tesseract` binary.
See the project README for the full workflow.
